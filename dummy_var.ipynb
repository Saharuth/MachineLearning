แปลงข้อความในคอลัมน์ให้เป็นตัวเลข เพื่อใช้ในการ train test model แบบ linear ได้

```python
import pandas as pd
print(f'pandas  version: {pd.__version__}')
```

```python
df = pd.read_csv('https://github.com/Saharuth/MachineLearning/raw/main/msleep.csv')
df.head()
```

```python
df.info()
# ดูข้อมูลเพื่อตรวจสอบว่า แต่ละคอลัมน์มี missing value กี่แถว
```

###Check NA in 'vore' column

```python
df[df.vore.isna()] # แสดงแถวที่มีคอลัมน์ vore เป็น NaN
```

```python
df.dropna(subset=['vore'], axis='index', inplace=True) # inplace => ให้กำหนดค่า df ที่กรองแล้ว ลงใน df ไหม
df
```

###Create Dummy Variables

```python
dummies = pd.get_dummies(df['vore'])
dummies
```

```python
dm = pd.concat([df, dummies], axis='columns')
dm.info()
```

###sklearn: LinearRegression

```python
import sklearn
from sklearn.linear_model import LinearRegression
print(f'sklearn version: {sklearn.__version__}')
```

```python
model = LinearRegression()
dt = dm[['vore', 'bodywt', 'carni', 'herbi', 'omni', 'insecti', 'sleep_total']].dropna(axis='index')
# ลบแถวที่มี NaN
X = dt[['bodywt', 'insecti', 'herbi', 'omni']]
```

```python
print(dt.info())
print(X.info())
```

```python
model.fit(X, dt.sleep_total)
model.score(X, dt.sleep_total) # R-squared
```

```python
model.coef_
```

###Statmodels: Linear Regression

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
dt.columns
```

```python
model_a = smf.ols(formula='sleep_total ~ bodywt + insecti + herbi + omni', data=dt).fit()
```

```python
model_a.pvalues # parameter values
```

```python
print(model_a.summary())
```

###use C() in Statsmodel to create dummy variables internally

```python
model_b = smf.ols(formula='sleep_total ~ bodywt + C(vore)', data=dt).fit()
# c function จะทำหน้าที่สร้าง dummy variables ภายในเองโดยอัตโนมัติ
```

```python
print(model_b.summary())
```
