###Impute/Replace missing values
#####วิธีการจัดการ missing values => string, number

```python
import pandas as pd
import numpy as np
print(f'pandas  version: {pd.__version__}')
print(f'numpy   version: {np.__version__}')
```

```python
df = pd.read_csv('https://github.com/Saharuth/MachineLearning/raw/main/msleep.csv')
df = df.sample(20, random_state=123)
df.info()
```

###check NA in 'vore' column

```python
df[df.vore.isna()]
```

###sklearn => SimpleImputer

```python
import sklearn
from sklearn.impute import SimpleImputer
```

```python
print(f'sklearn version: {sklearn.__version__}')
```

```python
imp = SimpleImputer(strategy='most_frequent') # เลือกแทนที่ค่าด้วยค่าที่มีความถี่มากที่สุด
imp.fit_transform(df[['vore']]) # ผ่านค่าแบบ dataframe เข้าไป
```

```python
df['vore2'] = imp.fit_transform(df[['vore']])
df[df.vore.isna()][['name', 'vore', 'vore2']]
```

```python
imp2 = SimpleImputer(strategy='constant', fill_value='omni') # กำหนดค่า missing ให้เป็น omni
df['vore3'] = imp2.fit_transform(df[['vore']])
df[df.vore.isna()][['name', 'vore', 'vore2', 'vore3']]
```

```python
# imp3 = SimpleImputer(strategy='mean') # แทนที่ตัวเลขที่ missing ด้วยค่า mean ของคอลัมน์นั้น
# imp3 = SimpleImputer(strategy='median')
imp3 = SimpleImputer(strategy='constant', fill_value=-99)
df['sleep_rem2'] = imp3.fit_transform(df[['sleep_rem']])
df[df.sleep_rem.isna()][['name', 'sleep_rem', 'sleep_rem2']]
```

```python
# ถ้า missing value เป็นตัวเลขที่เขากำหนดมา เช่น -99
imp_x = SimpleImputer(missing_values=-99, strategy='constant', fill_value=np.nan)
df['sleep_rem3'] = imp_x.fit_transform(df[['sleep_rem2']])
df[df.sleep_rem2 == -99.0][['name', 'sleep_rem', 'sleep_rem2', 'sleep_rem3']]
```
